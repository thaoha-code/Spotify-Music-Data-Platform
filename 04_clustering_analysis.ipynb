TOP 50 THỂ LOẠI CÓ NHIỀU NGHỆ SĨ HÁT NHẤT

```python
from collections import Counter
# Truy vấn dữ liệu từ bảng Artists
query = """
    SELECT genres
    FROM Artists
"""
genres_data = pd.read_sql(query, conn)

# Tách chuỗi thể loại nhạc và loại bỏ giá trị NULL hoặc trống
all_genres = []
for genres in genres_data['genres']:
    if genres:  # Kiểm tra nếu genres không phải NULL hoặc trống
        genre_list = genres.split(',')
        all_genres.extend([genre.strip() for genre in genre_list])

# Đếm số lượng nghệ sĩ cho mỗi thể loại nhạc
genre_counts = Counter(all_genres)

# Lấy 50 thể loại nhạc phổ biến nhất
top_50_genres = genre_counts.most_common(50)

# Chuyển dữ liệu thành DataFrame
top_50_genres_df = pd.DataFrame(top_50_genres, columns=['Genre', 'Num_Artists'])
print(top_50_genres_df)
```

```python
# Tạo màu sắc gradient từ xanh lá cây
colors = plt.cm.summer(np.linspace(0, 1, len(top_50_genres_df)))
# Vẽ biểu đồ
plt.figure(figsize=(14, 8))
plt.barh(top_50_genres_df['Genre'], top_50_genres_df['Num_Artists'], color=colors)
plt.gca().invert_yaxis()  # Đảo ngược trục Y để thể loại phổ biến nhất nằm trên cùng
plt.title('Top 50 Most Popular Music Genres by Number of Artists')
plt.xlabel('Number of Artists')
plt.ylabel('Genre')
plt.tight_layout()
plt.grid(False)  # Tắt lưới

plt.show()
```

```python
# Truy vấn dữ liệu thể loại nhạc từ cơ sở dữ liệu
query = """
    SELECT genres
    FROM Artists
"""
genres_data = pd.read_sql(query, conn)

# Tách chuỗi thể loại nhạc và tạo danh sách tất cả thể loại
all_genres = []
for genres in genres_data['genres']:
    # Tách chuỗi thể loại và loại bỏ dấu cách dư thừa
    genre_list = genres.split(',') 
    all_genres.extend([genre.strip() for genre in genre_list])

# Đếm số lượng từng thể loại nhạc
genre_counts = pd.Series(all_genres).value_counts()
print(genre_counts)
```

TOP 50 GENRE PHỔ BIẾN NHẤT

```python
# Lấy dữ liệu
artist_genres = pd.read_sql("""
    WITH GenreSplit AS (
        SELECT value AS genre, id AS artist_id
        FROM Artists
        CROSS APPLY STRING_SPLIT(genres, ',')
    )
    SELECT TOP 50 genre, COUNT(DISTINCT artist_id) AS num_artists
    FROM GenreSplit
    GROUP BY genre
    ORDER BY num_artists DESC
""", conn)
print(artist_genres)
```

```python
# Dữ liệu đã được xử lý từ trước
artist_genres_sorted = artist_genres.sort_values('num_artists', ascending=False)

# Tạo biểu đồ với seaborn
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Sử dụng palette màu xanh lá nhiều sắc độ (Greens) để làm nổi bật các thể loại
sns.barplot(x='genre', y='num_artists', data=artist_genres_sorted.head(50), palette='Greens')

# Thêm tiêu đề và nhãn trục
plt.title('Top 50 Most Popular Genres by Number of Artists', fontsize=18, fontweight='bold')
plt.xlabel('Genres', fontsize=14)
plt.ylabel('Number of Artists', fontsize=14)

# Định dạng lại các nhãn trục X và Y
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Cải thiện độ sắc nét của biểu đồ
plt.tight_layout()
plt.grid(False)  # Tắt lưới
plt.show()
```

TOP 50 NGHỆ SĨ CÓ LƯỢNG NGƯỜI THEO DỖI CAO NHẤT

```python
# Truy vấn dữ liệu Top 50 nghệ sĩ có số lượng người theo dõi cao nhất
top_artists = pd.read_sql("""
    SELECT TOP 50 name, num_followers
    FROM Artists
    ORDER BY num_followers DESC
""", conn)
print(top_artists)
```

```python
# Hàm định dạng trục hiển thị dưới dạng triệu (Million)
def millions_formatter(x, pos):
    return f'{x / 1e6:.1f}M'  # Chia giá trị cho 1e6 để chuyển đổi sang triệu

# Sắp xếp lại các nghệ sĩ theo số lượng followers
top_artists_sorted = top_artists.sort_values('num_followers', ascending=False)

# Tạo biểu đồ với seaborn
plt.figure(figsize=(14, 10))
sns.set(style="whitegrid")

# Sử dụng màu sắc gradient từ matplotlib (sắc độ xanh lá)
colors = plt.cm.Greens(np.linspace(1, 0, len(top_artists_sorted.head(50))))  # Đảo ngược gradient

# Vẽ biểu đồ với các màu sắc xanh lá đa dạng
sns.barplot(
    x='name', 
    y='num_followers', 
    data=top_artists_sorted.head(50), 
    palette=colors
)

# Thêm tiêu đề và nhãn trục
plt.title('Top 50 Artists with the Most Followers', fontsize=18, fontweight='bold')
plt.xlabel('Artist Name', fontsize=14)
plt.ylabel('Number of Followers ', fontsize=14)

# Áp dụng định dạng trục Y để hiển thị số liệu theo triệu
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))

# Đặt các nhãn trục X và Y sao cho dễ đọc hơn
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)

# Cải thiện độ sắc nét và hiển thị
plt.tight_layout()
plt.grid(False)  # Tắt lưới
plt.show()
```

TOP 50 NGHỆ SĨ CÓ SỐ LƯỢNG NGƯỜI NGHE HÀNG THÁNG CAO NHẤT

```python
# Truy vấn kết hợp thông tin từ bảng ArtistData và bảng artists_sp
artist_data_with_listeners = pd.read_sql("""
    SELECT top 50 ad.ArtistID, ad.ArtistName, a.monthly_listeners
    FROM ArtistData ad
    JOIN Artists a ON ad.ArtistID = a.id
    ORDER BY a.monthly_listeners DESC
    
""", conn)

print(artist_data_with_listeners)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Giả sử bạn đã có bảng artist_data_with_listeners với cột ArtistName và monthly_listeners
# Chuyển đổi 'monthly_listeners' thành triệu (Million)
artist_data_with_listeners['monthly_listeners_million'] = artist_data_with_listeners['monthly_listeners'] / 1e6

# Sắp xếp dữ liệu để các nghệ sĩ có listeners cao nhất đứng đầu
artist_data_with_listeners_sorted = artist_data_with_listeners.sort_values('monthly_listeners_million', ascending=False)

# Vẽ biểu đồ
plt.figure(figsize=(14, 10))
sns.set(style="whitegrid")

# Sử dụng màu sắc đẹp và dễ nhìn
sns.barplot(
    x='monthly_listeners_million', 
    y='ArtistName', 
    data=artist_data_with_listeners_sorted.head(50), 
    palette='viridis'
)

# Đảo ngược trục Y để nghệ sĩ có số lượng listeners cao nhất đứng đầu
plt.gca().invert_yaxis()

# Đặt tiêu đề và nhãn cho các trục
plt.title('Top 50 Artists with the Highest Monthly Listeners', fontsize=18, fontweight='bold')
plt.xlabel('Monthly Listeners (in Millions)', fontsize=14)
plt.ylabel('Artist Name', fontsize=14)

# Chỉnh sửa nhãn trục x để hiển thị số liệu theo triệu (million)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.1f}M'))

# Tăng độ sắc nét của các nhãn trục x và y
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Căn chỉnh và hiển thị biểu đồ
plt.tight_layout()
plt.grid(False)  # Tắt lưới
plt.show()
```

```python
# Truy vấn dữ liệu từ bảng audio_feature
query = """
SELECT 
    danceability, 
    energy, 
    speechiness, 
    acousticness, 
    instrumentalness, 
    liveness, 
    valence
FROM Audio_Features
"""
audio_features = pd.read_sql(query, conn)

# Hiển thị 5 dòng đầu tiên
print(audio_features.head())
```

```python
from math import pi

# Tính trung bình giá trị của mỗi đặc trưng (column)
audio_means = audio_features.mean()

# Chuẩn bị dữ liệu cho radar chart
categories = list(audio_means.index)  # Tên các cột
values = audio_means.values.tolist()  # Giá trị trung bình của mỗi cột
values += values[:1]  # Lặp lại giá trị đầu tiên để đóng vòng tròn

# Số lượng đặc trưng
num_vars = len(categories)

# Góc cho mỗi trục của radar
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

# Tạo radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Vẽ vùng dữ liệu
ax.fill(angles, values, color='green', alpha=0.25)
ax.plot(angles, values, color='green', linewidth=2)

# Gắn nhãn cho mỗi trục
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, color='darkgreen', fontsize=12)

# Gắn nhãn giá trị trục Y
ax.yaxis.set_tick_params(labelsize=10, color='white')
ax.set_yticklabels([])

# Thêm tiêu đề
plt.title('Average Audio Features', fontsize=16, color='darkgreen', weight='bold', pad=20)

# Hiển thị biểu đồ
plt.show()
```
